# **Veri Çoğaltma** İşlenmemiş veri setinde

```python
import pandas as pd
from sklearn.utils import resample

# Örnek olarak dengesiz bir veri kümesi yükleyelim
veri_seti = pd.read_excel("/content/CFTR_all_datas adlı dosyanın kopyası.xlsx")

# Öncelikle sınıflar arasındaki dengesizliği belirleyelim
sinif_0 = veri_seti[veri_seti['Change from previous version?'] == 0]
sinif_1 = veri_seti[veri_seti['Change from previous version?'] == 1]

# Daha az olan sınıfa göre örnekleme yaparak dengeleyelim
sinif_1_arttirilmis = resample(sinif_1, replace=True, n_samples=len(sinif_0), random_state=42)

# Yeni dengelemiş veri kümesini oluşturalım
denge_veri_seti = pd.concat([sinif_0,sinif_1, sinif_1_arttirilmis])

# Yeni veri kümesini dışa aktaralım
denge_veri_seti.to_excel("denge_veri_seti.xlsx", index=False)
```

```python

# Check the distribution of each class
print("Class 0 count:", len(sinif_0))
print("Class 1 count:", len(sinif_1))
```

## **'# alleles in CFTR2' Normalizasyon**

```python
veri_seti = pd.read_excel('/content/one_hot_encoded_veri_seti.xlsx')

# Normalizasyon uygulanacak sütun adı
sutun_adı = '# alleles in CFTR2'

# Min-Max normalizasyonu formülü: (x - min) / (max - min)
min_deger = veri_seti[sutun_adı].min()
max_deger = veri_seti[sutun_adı].max()

veri_seti[sutun_adı] = (veri_seti[sutun_adı] - min_deger) / (max_deger - min_deger)

# Normalized veriyi yeni bir Excel dosyasına yaz
veri_seti.to_excel('/content/CFTR_final.xlsx', index=False)
```
